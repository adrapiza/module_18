```python
#Import Libraries
import pandas as pd 
```

```python
# Set path and read files 
July_csv = 'JC-202407-citibike-tripdata.csv'
August_csv = 'JC-202408-citibike-tripdata.csv'
September_csv = 'JC-202409-citibike-tripdata.csv'

data_july = pd.read_csv(July_csv)
data_aug = pd.read_csv(August_csv)
data_sept = pd.read_csv(September_csv)
```

```python
#Print number of rows in the months and the first few rows
print("Number of rows in July:", data_july.shape[0])
print(data_july.head())
```

```python
print("Number of rows in August:", data_aug.shape[0])
print(data_aug.head())
```

```python
print("Number of rows in September:", data_sept.shape[0])
print(data_sept.head())
```

```python
# Check for missing values in each monthly dataset
missing_values_july = data_july.isnull().sum()
print("Missing values in July data:\n", missing_values_july)
```

```python
missing_values_aug = data_aug.isnull().sum()
print("\nMissing values in August data:\n", missing_values_aug)
```

```python
missing_values_sept = data_sept.isnull().sum()
print("\nMissing values in September data:\n", missing_values_sept)
```

```python
#Dropping rows with missing values 
 
data_july_1= data_july.dropna()
data_aug_1 = data_aug.dropna()
data_sept_1 = data_sept.dropna()
```

```python
# Check the number of rows before and after dropping missing values
print("Original July data rows:", len(data_july))
print("July data after dropping missing values:", len(data_july_1))

print("Original August data rows:", len(data_aug))
print("August data after dropping missing values:", len(data_aug_1))

print("Original September data rows:", len(data_sept))
print("September data after dropping missing values:", len(data_sept_1))
```

```python
# Check columns and data types in each cleaned DataFrame
print(data_july_1.dtypes)
```

```python
print(data_aug_1.dtypes)
```

```python
print(data_sept_1.dtypes)
```

```python
print("July Columns:", data_july_1.columns)
print("August Columns:", data_aug_1.columns)
print("September Columns:", data_sept_1.columns)
```

```python
# Concatenate the monthly data
combined_data = pd.concat([data_july_1, data_aug_1, data_sept_1], ignore_index=True)
print("Combined data rows:", len(combined_data))
```

```python
# Descriptive statistics
print(combined_data.describe())

print("User Types:", combined_data['member_casual'].unique())
print("Start Stations:", combined_data['start_station_name'].nunique())
print("End Stations:", combined_data['end_station_name'].nunique())
```

```python
print(combined_data)
```

```python
# Display dataframe
combined_data.head()
```

```python
#Convert DataFrame into CSV
# Export the cleaned DataFrame to a CSV file
combined_data.to_csv('cleaned_citi_bike_data.csv', index=False)
```

```python
#Peak times 
# Convert started_at to datetime
combined_data['started_at'] = pd.to_datetime(combined_data['started_at'])

# Extract the hour from the started_at column
combined_data['hour'] = combined_data['started_at'].dt.hour

# Group by hour and count the number of rides
peak_times = combined_data.groupby('hour').size().reset_index(name='ride_count')



# Sort the result to find peak hours
peak_times = peak_times.sort_values(by='ride_count', ascending=False)

#Reset Index 
peak_times.reset_index(drop=True, inplace=True)

peak_times.head(50)
```

```python
#Create Peak Time CSV 
peak_times.to_csv('Peak_times.csv', index=False)
```

```python
#Rideship growth
# Extract month from the started_at column
combined_data['month'] = combined_data['started_at'].dt.month

# Group by month and count the number of rides
monthly_ridership = combined_data.groupby('month').size().reset_index(name='ride_count')

monthly_ridership.head()
```

```python
#Create CSV for Rideship Growth
monthly_ridership.to_csv('Rideship Growth.csv', index=False)
```

```python
#Trip Duration 
# Convert 'ended_at' to datetime
combined_data['ended_at'] = pd.to_datetime(combined_data['ended_at'])

# Calculate trip duration in minutes
combined_data['duration'] = (combined_data['ended_at'] - combined_data['started_at']).dt.total_seconds() / 60

# Get summary statistics for trip duration
duration_stats = combined_data['duration'].describe()

# Convert the Series to a DataFrame
duration_stats_df = duration_stats.reset_index()

# Rename the columns for clarity
duration_stats_df.columns = ['Statistic', 'Value']

# Display the summary statistics DataFrame
duration_stats_df.head(10)

```

```python
# Create CSV for Trip Duration Stats
duration_stats_df.to_csv('Trip_Duration_Stats.csv', index=False)
```
